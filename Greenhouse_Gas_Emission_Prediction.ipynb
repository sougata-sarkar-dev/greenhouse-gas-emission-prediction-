```python
import pandas as pd

df = pd.read_csv('/content/SupplyChainEmissionFactorsforUSIndustriesCommodities2015_Summary.csv')
display(df.head())
display(df.info())
```

```python
df.dropna(subset=['Supply Chain Emission Factors with Margins'], inplace=True)
display(df.info())
```

```python
top_10_industries = df.groupby('Industry Name')['Supply Chain Emission Factors with Margins'].sum().sort_values(ascending=False).head(10).reset_index()
display(top_10_industries)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x='Industry Name', y='Supply Chain Emission Factors with Margins', data=top_10_industries)
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Industries by Supply Chain Emission Factors')
plt.xlabel('Industry Name')
plt.ylabel('Supply Chain Emission Factors with Margins')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Supply Chain Emission Factors with Margins'], kde=True, bins=20)
plt.title('Distribution of Supply Chain Emission Factors')
plt.xlabel('Supply Chain Emission Factors with Margins')
plt.ylabel('Frequency')
plt.show()
```
